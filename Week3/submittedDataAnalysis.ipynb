```python
import pickle
import matplotlib.pyplot as plt
import numpy as np
```

```python
def read(address):
    file = open(address,"rb")
    data = pickle.load(file)
    file.close()
    output, target, before = data
    output = output.detach().cpu().numpy()
    return output, target, before

def plot(output, target, before, title="Title"):
    plt.hist(output.flatten(),bins=10, alpha=0.5,label="Output")
    plt.hist(target.flatten(),bins=10, alpha=0.5,label="Target")
    plt.hist(before.flatten(),bins=10, alpha=0.5,label="Before")
    plt.legend()
    plt.title(title)
    plt.show()
```

```python
for i in range(0,10):
    output, target, before = read(f"Results/id_{i}.obj")
    plot(output, target, before,f"Layer {i}")
```
